```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create Blank Figure
fig = plt.figure(figsize=(10, 10))

# Pre-set size of data frame
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
```

```python


food_data = pd.read_csv("../../final_data_2/cleaned_indv_items.csv", index_col=0)
food_data.reset_index(inplace=True)
print(food_data.shape)
food_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
menu_data = food_data
# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Creating the plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Box Plot for Menu Prices
sns.boxplot(x=menu_data['price'], ax=axes[0, 0])
axes[0, 0].set_title('Box Plot of Menu Prices')
axes[0, 0].set_xlabel('Price ($)')
axes[0, 0].set_ylabel('Frequency')

# Histogram for Menu Prices
sns.histplot(menu_data['price'], bins=20, ax=axes[0, 1], kde=True)
axes[0, 1].set_title('Histogram of Menu Prices')
axes[0, 1].set_xlabel('Price ($)')
axes[0, 1].set_ylabel('Frequency')

# Box Plot for Calories per Dollar
sns.boxplot(x=(menu_data['Calories']/menu_data['price']), ax=axes[1, 0])
axes[1, 0].set_title('Box Plot of Calories per Dollar')
axes[1, 0].set_xlabel('Calories per Dollar')
axes[1, 0].set_ylabel('Frequency')

# Histogram for Calories per Dollar
sns.histplot((menu_data['Calories']/menu_data['price']), bins=20, ax=axes[1, 1], kde=True)
axes[1, 1].set_title('Histogram of Calories per Dollar')
axes[1, 1].set_xlabel('Calories per Dollar')
axes[1, 1].set_ylabel('Frequency')

# Adjusting layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Creating a 2x2 subplot structure
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Box Plot for Menu Prices
sns.boxplot(x=menu_data['price'], ax=axes[0, 0])
axes[0, 0].set_title('Box Plot of Menu Prices')
axes[0, 0].set_xlabel('Price ($)')

# Histogram for Menu Prices
sns.histplot(menu_data['price'], bins=20, ax=axes[0, 1], kde=True)
axes[0, 1].set_title('Histogram of Menu Prices')
axes[0, 1].set_xlabel('Price ($)')
axes[0, 1].set_ylabel('Count')

# Box Plot for Calories per Dollar
sns.boxplot(x=(menu_data['Calories']/menu_data['price']), ax=axes[1, 0])
axes[1, 0].set_title('Box Plot of Calories per Dollar')
axes[1, 0].set_xlabel('Calories per Dollar')

# Histogram for Calories per Dollar
sns.histplot((menu_data['Calories']/menu_data['price']), bins=20, ax=axes[1, 1], kde=True)
axes[1, 1].set_title('Histogram of Calories per Dollar')
axes[1, 1].set_xlabel('Calories per Dollar')
axes[1, 1].set_ylabel('Count')

# Adjusting layout for better spacing
plt.tight_layout()
plt.show()
```
