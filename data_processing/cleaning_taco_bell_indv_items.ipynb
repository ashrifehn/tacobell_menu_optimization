```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Reading in the data
indv_items = pd.read_csv("../data/uncleaned_taco_bell_indv_items.csv", index_col=0)
print(indv_items.shape)

indv_items.info()
```

```python
indv_items.head(30)
```

- As we can see all the NaN values are only for alleregen info, so i will be replacing all NaN with zero

```python
indv_items.fillna(0, inplace=True)

indv_items.info()
```

```python
def remove_substrings(df, substrings=['mcg', '$', 'mg', 'g', '<']):
    # Loop over columns and replace substrings
    for col in df.columns:
        if df[col].dtype == 'object':  # Check if column is of string type
            for substring in substrings:
                df[col] = df[col].str.replace(substring, '')
    return df
```

```python
indv_items_data = remove_substrings(indv_items)
indv_items_data.head(30)
```

```python
# Changing data types from object to float

print(f"Prevouse data types\n{indv_items_data.dtypes}")

def object_to_float(df):
# Convert the specified columns to float
    for col in df.columns[1:]:
        df[col] = df[col].astype(float)
    return df



indv_items_data = object_to_float(indv_items_data)
print(f"Updated data types: \n{indv_items_data.dtypes}")
```

- If price is less than $1.00 that means it is a sauce packet and will get dropped from the data frame

```python
indv_items_data = indv_items_data[indv_items_data['price'] >= 1.00]
indv_items_data.shape
```
